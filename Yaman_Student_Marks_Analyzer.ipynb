PROJECT TITLE :- Student Marks Analyzer using Python and NumPy  
PROBLEM STATEMENT :- Creating a python program that collects student marks analyzes them to find total marks, average marks, highest and lowest marks, and gives a grade based on performance.


```python
# Import NumPy library to work with arrays
import numpy as np


# Function to analyze student marks
def analyze_marks(marks):

    # Convert the list of marks into a NumPy array
    marks_array = np.array(marks)

    # Calculate total marks using NumPy
    total = np.sum(marks_array)

    # Calculate average marks
    average = np.mean(marks_array)

    # Find highest marks
    highest = np.max(marks_array)

    # Find lowest marks
    lowest = np.min(marks_array)

    # Assign grade based on average marks
    if average > 85:
        grade = "A"
    elif average >= 70:
        grade = "B"
    elif average >= 50:
        grade = "C"
    else:
        grade = "Fail"

    # Return all calculated values
    return total, average, highest, lowest, grade


# Ask user how many students
n = int(input("Enter number of students: "))

# Create empty list to store marks
marks_list = []

# Loop to take marks one by one
for i in range(n):
    mark = float(input(f"Enter marks for student {i+1}: "))
    marks_list.append(mark)

# Call function and store returned values
total, average, highest, lowest, grade = analyze_marks(marks_list)

# Display the results
print("\n----- RESULT ANALYSIS -----")
print("Marks Entered:", marks_list)
print("Total Marks:", total)
print("Average Marks:", average)
print("Highest Marks:", highest)
print("Lowest Marks:", lowest)
print("Grade:", grade)
```
